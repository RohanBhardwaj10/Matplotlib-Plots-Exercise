**SUBPLOTS**

```python
import numpy as np
import matplotlib.pyplot as plt
c=np.linspace(0,10,11)
d=c**4
```

```python
x=np.arange(0,10)
y=2*x
```

```python
fig,axes = plt.subplots()
```

```python
fig,axes = plt.subplots(nrows=1 , ncols=2)
axes[0].plot(x,y)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2 , ncols=2,  dpi=100)
axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].plot(x,y)
axes[1][1].plot(c,d)
fig.subplots_adjust(wspace=0.9,hspace=0.2,top=None , bottom=None , left=None , right=None)
fig.suptitle("FIGURE LEVEL")
fig.savefig("figlevel.png",bbox_inches='tight')
```
